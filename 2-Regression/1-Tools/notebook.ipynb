```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Load the [diabetes dataset](https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset) and print out the first entry


```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
print(y[0])
```

Select the column with index 2 as the feature 

```python
X = X[:, np.newaxis, 0]
print(X.shape)
print(X[0])
```

Split the training and test data for both `X` and `y`

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

Select the model and fit it with the training data

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Create a prediction using test data

```python
y_pred = model.predict(X_test)
```

Show test data in a scatterplot and the predicted data as a line 

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
